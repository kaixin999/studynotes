```python
x = np.arange(0,9,0.1)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y,label='sin')
plt.title = 'sin and cos'
plt.plot(x,y2,linestyle='--',label='cos')
plt.legend()
plt.show()
```
